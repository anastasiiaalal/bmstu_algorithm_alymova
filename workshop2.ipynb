# Задача 1

```python
def get_max_substr_len(s):
    dct = {}
    start = 0
    res = 0

    for i in range(len(s)):
        if s[i] in dct and dct[s[i]] >= start:
            start = dct[s[i]] + 1
            
        dct[s[i]] = i
        
        if i - start + 1 > res:
            res = i - start + 1

    return res

get_max_substr_len('ABCDEBCFGH')
```

# Задача 2

```python
def get_k_elem(m, k):
    n = len(m)
    mini, maxi = m[0][0], m[n - 1][n - 1]

    while mini < maxi:
        mid_elem = (mini + maxi) // 2

        count_less_elem = 0
        i, j = n - 1, 0

        while i >= 0 and j < n:
            if m[i][j] <= mid_elem:
                count_less_elem += i + 1
                j += 1
            else:
                i -= 1
        
        if count_less_elem >= k:
            maxi = mid_elem
        else:
            mini = mid_elem + 1

    return mini

get_k_elem([
    [1, 5, 9, 11],
    [2, 6, 10, 12],
    [3, 7, 13, 14],
    [4, 8, 15, 16]
], 8)
```
