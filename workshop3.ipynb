# Задача 1

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def level_order(root):
    if not root:
        return []

    res = []
    q = [root]
    i = 0

    while i < len(q):
        level_len = len(q) - i
        level = []

        for _ in range(level_len):
            node = q[i]
            i += 1
            level.append(node.val)
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)

        res.append(level)

    return res


def build_tree_iterative(arr):
    if not arr or arr[0] is None:
        return None

    nodes = [TreeNode(arr[0])]
    root = nodes[0]
    j = 0

    for k in range(1, len(arr), 2):
        parent = nodes[j]
        j += 1

        if k < len(arr) and arr[k] is not None:
            parent.left = TreeNode(arr[k])
            nodes.append(parent.left)

        if k + 1 < len(arr) and arr[k + 1] is not None:
            parent.right = TreeNode(arr[k + 1])
            nodes.append(parent.right)

    return root

root1 = build_tree_iterative([9, 16, 8, 6, 11, 10])
print(level_order(root1))  # [[9], [16, 8], [6, 11, 10]]

root2 = build_tree_iterative([1])
print(level_order(root2))  # [[1]]

root3 = build_tree_iterative([])
print(level_order(root3))  # []
```

# Задача 2

```python
def findKthLargest(nums, k):
    heap = nums[:k]

    def sift_down(a, i, n):
        while True:
            l = 2 * i + 1
            r = l + 1
            m = i
            if l < n and a[l] < a[m]:
                m = l
            if r < n and a[r] < a[m]:
                m = r
            if m == i:
                break
            a[i], a[m] = a[m], a[i]
            i = m

    def heapify(a):
        n = len(a)
        for i in range(n // 2 - 1, -1, -1):
            sift_down(a, i, n)

    heapify(heap)

    for x in nums[k:]:
        if x > heap[0]:
            heap[0] = x
            sift_down(heap, 0, k)

    return heap[0]


print(findKthLargest([3,2,1,5,6,4], 2))        # 5
print(findKthLargest([3,2,3,1,2,4,5,5,6], 4))  # 4
```
