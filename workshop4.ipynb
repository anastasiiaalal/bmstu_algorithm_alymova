# Задача 1

```python
def seq_length(nums):
    n = len(nums)
    dp = [1] * n
    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
```

```python
seq_length([5, 1, 6, 2, 7, 3, 8, 4]) # 4 - 1,2,3,4 или 5,6,7,8
```

```python
seq_length([4, 3, 2, 1]) # 1
```

```python
seq_length([10, 9, 2, 5, 3, 7, 101, 18]) # 4 
```

# Задача 2

```python
from collections import deque

def calc(matrix):
    if not matrix or not matrix[0]:
        return []

    m = len(matrix)
    n = len(matrix[0])

    res = [[10**8 for _ in range(n)] for _ in range(m)]
    q = deque()

    for i in range(m):
        for j in range(n):
            if matrix[i][j] == 0:
                res[i][j] = 0
                q.append((i, j))

    while q:
        i, j = q.popleft()

        for di, dj in ((1,0), (-1,0), (0,1), (0,-1)):
            ni = i + di
            nj = j + dj

            if 0 <= ni < m and 0 <= nj < n:
                if res[ni][nj] > res[i][j] + 1:
                    res[ni][nj] = res[i][j] + 1
                    q.append((ni, nj))

    return res
```

```python
calc(
        [[0,0,0],
         [0,1,0],
         [1,1,1]]
    )
```
