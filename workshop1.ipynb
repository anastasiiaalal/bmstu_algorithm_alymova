### Задача 1

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def find_intersection(headA, headB):
    n = 0
    a = headA
    while a:
        n += 1
        a = a.next

    m = 0
    b = headB
    while b:
        m += 1
        b = b.next

    diff = abs(n - m)

    if n > m:
        headA, headB = headB, headA

    for _ in range(diff):
        headB = headB.next

    a = headA
    b = headB

    while a != b:
        a = a.next
        b = b.next

    return a.val if a else None

common = ListNode(1)
common.next = ListNode(8)
common.next.next = ListNode(4)
common.next.next.next = ListNode(5)
headA = ListNode(4)
headA.next = common
headB = ListNode(5)
headB.next = ListNode(6)
headB.next.next = common

value = find_intersection(headA, headB)
print(value)

headC = ListNode(1)
headC.next = ListNode(2)
headD = ListNode(3)
headD.next = ListNode(4)

value = find_intersection(headC, headD)
print(value)
```

### Задача 2

```python
def sort_stack(stack):
    dop_stack = []
    while stack:
        x = stack.pop()

        while dop_stack and dop_stack[-1] > x:
            stack.append(dop_stack.pop())

        dop_stack.append(x)

    while dop_stack:
        stack.append(dop_stack.pop())

    return stack

stack = [3, 2, 1, 4, 7, 8]
sort_stack(stack)
```
